```
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
import math
headers = ["ID","CT","UCSize","UCShape","MA","SECSize","BN","BC","NN","Mitoses","Diagnosis"]
data = pd.read_csv('breast-cancer-wisconsin.data', na_values='?',    
         header=None, index_col=['ID'], names = headers) 
data = data.reset_index(drop=True)
data = data.fillna(0)
y=data.iloc[:,9]
y=np.array(y)
#y=y.reshape(-1,1)
X=data.drop("Diagnosis",1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
X_train=np.array(X_train)
X_test=np.array(X_test)
#print(y[0])
```

```
class Node:
  def __init__(self,num_samples,totalcnt,prediction):
    self.num_samples = num_samples
    self.totalcnt=totalcnt
    self.prediction=prediction
    self.feature_index = 0
    self.threshold = 0
    self.left = None
    self.right = None
```

```
class treeClassifier:
  def __init__(self,md,mode):
    self.mode=mode
    self.max_depth=md
  def expand_tree(self,X,y,depth):
    totalcnt=[(y==2).sum(),(y==4).sum()]
    prediction=-1
    if(totalcnt[0]>totalcnt[1]):
      prediction=2
    else:
      prediction=4
    if(self.mode=="gini"):
      gini=self.ginival(totalcnt[0],totalcnt[1])
    elif(self.mode=="entropy"):
      entropy=self.entropyval(totalcnt[0],totalcnt[1])
    node = Node(len(y),totalcnt,prediction)
    if(self.mode=="gini" and gini==0):
      return node
    elif(self.mode=="entropy" and entropy==0):
      return node
    if depth < self.max_depth:
      if(self.mode=="gini"):
        bestidx,bestgini,bestth=self.best_split(X,y,9)
      elif(self.mode=="entropy"):
        bestidx,bestentro,bestth=self.best_split(X,y,9)
      if bestidx!=-1:
        indices_right = X[:, bestidx]>bestth
        indices_left = X[:, bestidx]<=bestth
        X_right, y_right=X[indices_right], y[indices_right]
        X_left, y_left=X[indices_left], y[indices_left]
        node.feature_index = bestidx
        node.threshold = bestth
        print(bestidx,bestth)
        node.left=self.expand_tree(X_left, y_left, depth + 1)
        node.right=self.expand_tree(X_right, y_right, depth + 1)
    return node
  def fit_model(self, X, y):
    self.n_features_ = 9
    self.n_classes_ = 2  
    self.Tree=self.expand_tree(X,y,0)
  def predict(self,X_test):
    ans=[]
    for i in X_test:
      node = self.Tree
      while node:
        if i[node.feature_index]<=node.threshold:
          if node.left!=None:
            node = node.left
          else:
            ans.append(node.prediction)
            break
        else:
          if node.right!=None:
            node=node.right
          else:
            ans.append(node.prediction)
            break
    #print(ans)
    return ans
  def ginival(self,a,b):
    temp=a+b
    if(temp==0):
      return 0
    return 1-(a**2+b**2)/(temp**2)
  def entropyval(self,a,b):
    temp=a+b
    if(temp==0):
      return 0
    elif(a==0 or b==0):
      return 0
    else:
      return -(a/temp)*math.log2(a/temp)-(b/temp)*math.log2(b/temp)

  def best_split(self,X,y,nf):
    bestidx=-1
    bestgini=1
    bestentro=1
    bestth=-1
    for i in range(nf):
      temp=[]
      cnt2=0
      cnt4=0
      for j in range(len(y)):
        if(y[j]==2):
          cnt2+=1
        else:
          cnt4+=1
        temp.append([X[j,i],y[j]])
      temp.sort(key=lambda x:x[0])
      k=0
      rightcnt=[cnt2,cnt4]
      leftcnt=[0,0]
      while(k<len(y)):
        th=temp[k][0]
        while(k<len(y) and temp[k][0]==th):
          if(temp[k][1]==2):
            rightcnt[0]-=1
            leftcnt[0]+=1
          if(temp[k][1]==4):
            rightcnt[1]-=1
            leftcnt[1]+=1
          k+=1
        #print(leftcnt,rightcnt,i)
        if(self.mode=="gini"):
          curginileft=self.ginival(leftcnt[0],leftcnt[1])
          curginiright=self.ginival(rightcnt[0],rightcnt[1])
          sum1=sum(leftcnt)
          sum2=sum(rightcnt)
          fingini=(sum1*curginileft+sum2*curginiright)/(sum1+sum2)
          if(fingini<bestgini):
            bestgini=fingini
            bestidx=i
            bestth=th
        elif(self.mode=="entropy"):
          curentroleft=self.entropyval(leftcnt[0],leftcnt[1])
          curentroright=self.entropyval(rightcnt[0],rightcnt[1])
          sum1=sum(leftcnt)
          sum2=sum(rightcnt)
          finentro=(sum1*curentroleft+sum2*curentroright)/(sum1+sum2)
          if(finentro<bestentro):
            bestentro=finentro
            bestidx=i
            bestth=th
    if(self.mode=="gini"):
      return bestidx,bestgini,bestth      
    elif(self.mode=="entropy"):
      return bestidx,bestentro,bestth
```

```
obj=treeClassifier(5,"entropy")
obj.fit_model(X_train,y_train)
y_pred=obj.predict(X_test)
print(y_pred)
num_correct = np.sum(y_pred == y_test)
accuracy = float(num_correct) / len(X_test)
print('Got %d / %d correct => accuracy: %f' % (num_correct, len(X_test), accuracy))
```
